### 다시 훈련 시작. 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# 이미지 다시 불러오기
iris = load_iris()

# 데이터를 훈련/테스트로 분리
features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    test_size=0.2, random_state=13)

# 훈련용/테스트용 잘 분리됐는지 확인
np.unique(y_test, return_counts=True)
```

```python
# 각 클래스(setosa, versicolor, verginica) 별로 동일 비율이 아니어서 이를 맞춰서 다시한번!
X_train, X_test, y_train, y_test = train_test_split(features, labels, stratify=labels,
                                                    test_size=0.2, random_state=13)

np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

# train 데이터로만 결정나무 모델 만들기
iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)

# train 데이터에 대한 accuracy 확인
y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))  # 더 큰 크기로 확대
_ = tree.plot_tree(iris_tree, feature_names=["length", "width"],
                   class_names=list(iris.target_names), filled=True, fontsize=10)
```

```python
from mlxtend.plotting import plot_decision_regions

# 훈련 데이터에 대한 결정 경계 확인
plt.figure(figsize=(12,6))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree, legend=2)
plt.show()
```

```python
# 테스트 데이터에 대한 정확도
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
from mlxtend.plotting import plot_decision_regions

# 테스트 데이터에 대한 결정 경계 확인
plt.figure(figsize=(12,6))
plot_decision_regions(X=X_test, y=y_pred_test, clf=iris_tree, legend=2)
plt.show()
```

```python
# 전체 데이터에서 관찰해보기
scatter_highlight_kwargs = {'s': 150, 'label': 'Test data', 'alpha': 0.9}
scatter_kwargs = {'s': 120, 'edgecolor': None, 'alpha': 0.7}

plt.figure(figsize=(12, 8))
plot_decision_regions(
    X=features, 
    y=labels,
    clf=iris_tree, 
    legend=2,
    X_highlight=X_test,
    scatter_highlight_kwargs=scatter_highlight_kwargs,
    scatter_kwargs=scatter_kwargs,
    contourf_kwargs={'alpha': 0.2}
)

plt.show()
```

```python
# 전체 특성을 사용한 결정나무 모델
fig = plt.figure(figsize=(8,6))
_ = tree.plot_tree(iris_tree, feature_names=list(iris.feature_names),
                   class_names=list(iris.target_names), filled=True, fontsize=10)
```

```python
# 모델을 사용하는 방법
# 예를들면, 길가다 주운 iris가 sepal과 petal의 length, width가 각각 [4.3, 2. , 1.2, 1.0]이라면?
# 전체 sepal + petal 사용해야함. 이전에는 iris_tree.fit(iris.data[:, 2:], iris.target) 했어서 다시 선언.

iris_tree.fit(iris.data, iris.target)  
test_data = [[4.3, 2.0, 1.2, 1.0]]
iris_tree.predict_proba(test_data)
```

```python
# 각 클래스별 확률이 아니라 범주 값을 바로 알고 싶다면

iris_tree.fit(iris.data, iris.target)  
test_data = [[4.3, 2.0, 1.2, 1.0]]
iris.target_names[iris_tree.predict(test_data)] # 이런식으로 하면 됨
```
