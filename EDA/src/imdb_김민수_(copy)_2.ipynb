```python
def get_details(url):
    details_list = []
    details_list_tmp = []
    writers_list = []

    req = Request(url=url, headers={'User-Agent':generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")

    writers_tmp = soup.find("ul", class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")

    for each in writers_tmp.find_all("div", class_="ipc-metadata-list-item__content-container"):
        for each_2 in each.find_all("li", class_="ipc-inline-list__item"):
            details_list_tmp.append(each_2.find("a", class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").get_text())
        tmp = [", ".join(details_list_tmp)]
        details_list.append(tmp)
        details_list_tmp = []
    
    return details_list
```

```python
import pandas as pd
import re

# 시즌, 에피소드, 이름 추출 함수
def extract_season_episode(name):
    match = re.search(r'S(\d+)\.E(\d+)\s*[∙·\-\–]?\s*(.+)', name)
    if match:
        season = int(match.group(1))
        episode = int(match.group(2))
        episode_name = match.group(3).strip()  # 에피소드 제목만 추출
        return season, episode, episode_name
    return None, None, name  # 매칭되지 않으면 기존 값 유지

# 데이터 가공
df[['Season', 'Episode', 'Name']] = df['Season_Episode_Name'].apply(lambda x: pd.Series(extract_season_episode(x)))

# 불필요한 컬럼 제거
df = df.drop(columns=['Season_Episode_Name', 'Link'])

# 리스트 형태(`[]`) 제거
df['Director'] = df['Director'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
df['Writers'] = df['Writers'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
df['Stars'] = df['Stars'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# 시즌과 에피소드 기준 정렬
df = df.sort_values(by=['Season', 'Episode']).reset_index(drop=True)

# 데이터 확인
df
```

```python
# 원하는 컬럼 순서로 재배치
df = df[['Season', 'Episode', 'Name', 'Rating', 'Story', 'Director', 'Writers', 'Stars']]

# 데이터 확인
df
```

```python
import matplotlib.pyplot as plt

# Rating 컬럼을 float로 변환
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")

# 시즌별 평균 평점 계산
season_avg_rating = df.groupby("Season")["Rating"].mean()

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(season_avg_rating.index, season_avg_rating.values, marker='o', linestyle='-', color='b')

# 그래프 스타일 설정
plt.xlabel("Season")
plt.ylabel("Average Rating")
plt.title("Average Episode Rating per Season")
plt.xticks(season_avg_rating.index)  # 시즌 라벨 맞춤
plt.grid(True)

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# ✅ 1️⃣ 감독별 평균 평점 비교
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")  # Rating을 숫자로 변환
director_avg_rating = df.groupby("Director")["Rating"].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
director_avg_rating.plot(kind="bar", color="skyblue")

plt.xlabel("Director")
plt.ylabel("Average Rating")
plt.title("Average Episode Rating by Director")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y")

plt.show()
```

```python
# 시즌별 에피소드 개수 계산
season_episode_count = df["Season"].value_counts().sort_index()

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.bar(season_episode_count.index, season_episode_count.values, color="lightcoral")

# 그래프 스타일 설정
plt.xlabel("Season")
plt.ylabel("Number of Episodes")
plt.title("Number of Episodes per Season")
plt.xticks(season_episode_count.index)
plt.grid(axis="y")

plt.show()
```

```python
# ✅ 가장 높은 평점을 받은 에피소드 TOP 10 시각화

# 가장 높은 평점을 받은 TOP 10 에피소드
top_10_episodes = df.nlargest(10, "Rating")[["Season", "Episode", "Name", "Rating"]]

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.barh(top_10_episodes["Name"], top_10_episodes["Rating"], color="goldenrod")

# 그래프 스타일 설정
plt.xlabel("Rating")
plt.ylabel("Episode Name")
plt.title("Top 10 Highest Rated Episodes")
plt.gca().invert_yaxis()  # 높은 평점이 위로 오도록 설정
plt.grid(axis="x")

# 그래프 표시
plt.show()
```
