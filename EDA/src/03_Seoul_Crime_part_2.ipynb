```python
import pandas as pd
import numpy as np

crime_raw_data = pd.read_csv(file_list[0], usecols=[0, 1, 2, 3], encoding="euc-kr")

crime_raw_data["죄종"] = crime_raw_data["죄종"].replace("강간,추행", "강간")

crime_station = crime_raw_data.pivot_table(
    crime_raw_data, index=["구분"], columns=["죄종", "발생검거"], aggfunc="sum"
)

crime_station.columns = [
    "강간검거",
    "강간",
    "강도검거",
    "강도",
    "살인검거",
    "살인",
    "절도검거",
    "절도",
    "폭력검거",
    "폭력"
]

crime_station["year"] = 2000
crime_station.head()
```

```python
crime_station_list = []

for each_file in file_list:
    print("file reading... ", each_file)

    crime_raw_data = pd.read_csv(each_file, usecols=[0, 1, 2, 3], encoding="euc-kr")

    crime_raw_data["죄종"] = crime_raw_data["죄종"].replace("강간,추행", "강간")

    crime_station = crime_raw_data.pivot_table(
        crime_raw_data, index=["구분"], columns=["죄종", "발생검거"], aggfunc="sum"
    )

    crime_station.columns = [
        "강간검거",
        "강간",
        "강도검거",
        "강도",
        "살인검거",
        "살인",
        "절도검거",
        "절도",
        "폭력검거",
        "폭력",
    ]

    crime_station["year"] = each_file.split("/")[-1][:4]

    crime_station.reset_index(inplace=True)
    crime_station_list.append(crime_station)
```

```python
crime_total = pd.concat(crime_station_list)
crime_total.head()
```

```python
crime_total_murder = crime_total.pivot(index="year", columns="구분", values="살인")
crime_total_murder
```

```python
f, ax = plt.subplots(2, 3, figsize=(24,12))

crime_types = ["강간", "강도", "살인", "절도", "폭력", "강간검거"]
districts = ["강남", "서초", "영등포", "노원", "관악"]

for i, crime in enumerate(crime_types):
    row = int(i/3)
    col = int(i%3)

    for district in districts:
        ax[row, col].plot(
            crime_total.pivot(index="year", columns="구분", values=crime)[district],
            label = district                          
        )

    ax[row, col].set_title("강남, 서초, 영등포, 노원, 관악의 " + crime +"의 변화")
    ax[row, col].legend(loc="upper right")

plt.tight_layout
plt.show()
```
