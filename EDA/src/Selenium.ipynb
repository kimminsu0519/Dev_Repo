```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from tqdm import tqdm
from glob import glob
import os

path = "../data/지역_*.xls"

for idx, gu_name in tqdm(enumerate(gu_names), total=len(gu_names)):
    gu = driver.find_element(By.ID, "SIGUNGU_NM0")
    gu.send_keys(gu_name)
    driver.implicitly_wait(40)
    driver.refresh()

    save = WebDriverWait(driver, 40).until(EC.element_to_be_clickable((By.CLASS_NAME, "btn_type6_ex_save")))
    # save.click()
    ActionChains(driver).move_to_element(save).perform()
    save.send_keys(Keys.ENTER)

    default_filename = os.path.join("/home/kms/dev_ws/eda/data", "지역_위치별(주유소).xls")  # 기본 다운로드 파일명
    new_filename = os.path.join("/home/kms/dev_ws/eda/data", f"지역_위치별(주유소) {gu_name}.xls")  # 변경할 파일명

    while (len(glob(path)) < idx + 1):
        time.sleep(0.1)

    os.rename(default_filename, new_filename)

len(glob(path))
```
