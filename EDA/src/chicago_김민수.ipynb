```python
import folium

df = pd.read_csv("../src/Chicago_Best_Restaurants.csv", index_col=0)
df.tail(10)
```

```python
lat_list = []
lng_list = []

for idx, row in tqdm(df.iterrows(), total=len(df.index)):
    if "|" in row["Address"]:
        lat_list.append(np.nan)
        lng_list.append(np.nan)
        continue

    geocode = gmaps.geocode(row["Address"])

    if len(geocode) == 0:
        lat_list.append(np.nan)
        lng_list.append(np.nan)
    else:
        lat = geocode[0].get("geometry")["location"]["lat"]
        lng = geocode[0].get("geometry")["location"]["lng"]
        
        lat_list.append(lat)
        lng_list.append(lng)

len(lat_list), len(lng_list)
```

```python
df["lat"] = lat_list
df["lng"] = lng_list

df.head
```

```python
df
```

```python
mapping = folium.Map(location=[lat_list[0], lng_list[0]], zoom_start=11)
mapping
```

```python
for idx, row in df.iterrows():
    if row["lat"] == np.nan or row["lng"] == np.nan:
        continue
    else:
        folium.Marker([row["lat"], row["lng"]], popup=row["Restaurant_Name"]).add_to(mapping)

mapping
```

```python
for idx, row in df.iterrows():
    if pd.isna(row["lat"]) or pd.isna(row["lng"]):  # NaN 값이 있으면 건너뛰기
        continue
    else:
        folium.Marker([row["lat"], row["lng"]], popup=row["Restaurant_Name"]).add_to(mapping)

mapping
```
