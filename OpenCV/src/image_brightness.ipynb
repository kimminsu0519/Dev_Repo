### 그레이스케일 영상 밝기 조절

```python
import numpy as np
import cv2
import sys

src = cv2.imread("../data/lenna.bmp", cv2.IMREAD_GRAYSCALE)

dst1 = cv2.add(src, 100)

dst2 = src + 100

cv2.imshow("src", src)
cv2.imshow("dst1", dst1)
cv2.imshow("dst2", dst2)

cv2.waitKey()
cv2.destroyAllWindows()
```

### 칼라영상 밝기 조절

```python
import numpy as np
import cv2
import sys

src = cv2.imread("../data/lenna.bmp")

dst1 = cv2.add(src, (100, 100, 100, 0))

dst2 = np.clip(src + 100, 0, 255).astype(np.uint8)

cv2.imshow("src", src)
cv2.imshow("dst1", dst1)
cv2.imshow("dst2", dst2)

cv2.waitKey()
cv2.destroyAllWindows()
```

### 덧셈 연산

```python
import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt

src1 = cv2.imread("../data/lenna.bmp")
src2 = cv2.imread("../data/sky.png")

resize_src2 = cv2.resize(src2, (src1.shape[1], src1.shape[0]))
dst = cv2.addWeighted(src1=src1, alpha=0.5, src2=resize_src2, beta=0.5, gamma=0)
show_dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(show_dst);
```

### 뺄셈 연산

```python
src1 = cv2.imread("../data/lenna.bmp", cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread("../data/circle.jpg", cv2.IMREAD_GRAYSCALE)

resize_src2 = cv2.resize(src2, (src1.shape[1], src1.shape[0]))
dst = cv2.subtract(src1, resize_src2)

plt.imshow(dst, cmap="gray");
```

### absdiff 함수

```python
img1 = cv2.imread("../data/road1.jpg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("../data/road2.jpg", cv2.IMREAD_GRAYSCALE)

dst = cv2.absdiff(img1, img2)

fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1, 3, 1)
plt.imshow(img1, cmap="gray")

fig.add_subplot(1, 3, 2)
plt.imshow(img2, cmap="gray")

fig.add_subplot(1, 3, 3)
plt.imshow(dst, cmap="gray")
```

### 명암비 조절

```python
src = cv2.imread("../data/lenna.bmp", cv2.IMREAD_GRAYSCALE)

if src is None:
    print("Image is failed")
    sys.exit()

alpha = 1.0

dst = np.clip((1+alpha)*src - 128*alpha, 0, 255).astype(np.uint8)

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

### 정규화

```python
src = cv2.imread("../data/Hawkes.jpg", cv2.IMREAD_GRAYSCALE)

if src is None:
    print("Image is failed")
    sys.exit()

dst = cv2.normalize(src, None, 0, 255, cv2.NORM_MINMAX)

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

### 컬러 영상의 히스토그램 평활화

```python
src = cv2.imread("../data/field.bmp")

if src is None:
    print("Image is failed")
    sys.exit()

src_ycrcb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)
ycrcb_planes = list(cv2.split(src_ycrcb))  # ✅ tuple → list

# 밝기 성분(Y)에만 히스토그램 평활화 적용
ycrcb_planes[0] = cv2.equalizeHist(ycrcb_planes[0])

# 병합 후 다시 BGR로 변환
dst_ycrcb = cv2.merge(ycrcb_planes)
dst = cv2.cvtColor(dst_ycrcb, cv2.COLOR_YCrCb2BGR)

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
