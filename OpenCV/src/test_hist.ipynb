```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.full((100,100,3), 255, dtype=np.uint8)
```

```python
a.shape
```

```python
a[:,:,0] = 0
a[:,:,1] = 0

plt.imshow(a)
```

```python
a_cvt = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
cv2.imwrite('a.jpg', a_cvt)
```

```python
a_ret = cv2.imread('./a.jpg')
```

```python
histvalue = cv2.calcHist([a_ret], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(histvalue)
```

```python
histvalue = cv2.calcHist([a_ret], channels=[2], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(histvalue)
```
