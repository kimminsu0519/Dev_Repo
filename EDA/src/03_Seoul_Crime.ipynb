```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

```python
crime_gu_norm_sort = crime_gu_norm.sort_values(by="검거", ascending=False)
crime_gu_norm_sort.head()
```

```python
target_col = [
    "성범죄검거율",
    "강도검거율",
    "살인검거율",
    "절도검거율",
    "폭력검거율",
    "검거",
]

plt.figure(figsize=(10, 10))
sns.heatmap(
    crime_gu_norm_sort[target_col], annot=True, fmt="f", linewidths=0.5, cmap="RdPu"
)
plt.title("범죄 검거 비율")
plt.show()
```

```python
target_col = ["성범죄", "강도", "살인", "절도", "폭력", "범죄"]
crime_gu_norm_sort = crime_gu_norm_sort.sort_values(by="범죄", ascending=False)

plt.figure(figsize=(10,10))
sns.heatmap(
    crime_gu_norm_sort[target_col], annot=True, fmt="f", linewidths=0.5, cmap="RdPu"
)
plt.title("범죄 발생 비율")
plt.show()
```
